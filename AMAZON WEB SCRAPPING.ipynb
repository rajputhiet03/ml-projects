```python
from selenium import webdriver
from bs4 import BeautifulSoup
import random
import os
import urllib
import urllib.request


```

```python
browser=webdriver.Chrome("D://driver//chromedriver.exe")
```

# FOR SAREES

```python
browser.get("https://www.amazon.in/s?k=sarees+for+women+latest+design+2020&crid=RQ8P1QA9CAQ2&sprefix=sarees%2Caps%2C459&ref=nb_sb_ss_i_1_6")
```

```python
html = browser.page_source
soup = BeautifulSoup(html)
```

```python
all_images = []
for link in soup.find_all('div',{"data-component-type":"s-search-result"},limit=1000):
    href = link.img['src']
    print(href)
    
    img_name = random.randrange(1000,2000)
    
    full_name = str(img_name) + ".jpg"
    
    urllib.request.urlretrieve(href, full_name)
    
    print("loop break")
```
