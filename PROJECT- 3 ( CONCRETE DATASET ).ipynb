```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(data))
z
```

```python
threshold=3
```

```python
print(np.where(z>3))
```

```python
z[42][1]
```

```python
data_new=data[(z<3).all(axis=1)]
data_new
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
data_new=scaler.fit_transform(data_new.astype(float))
data_new
```

```python
df=pd.DataFrame(data_new)
df.head()
```

# SPLITTING INTO x AND y

```python
x=df.drop(8,axis=1)
x.head()
```

```python
y=df[8]
y.head()
```

```python
x.shape
```

```python
y.shape
```

# TRAINING AND TESTING DATA

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.27,random_state=32)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# APPLYING MODELS

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_train,y_train)
```

```python
predlr=lr.predict(x_test)
```

```python
print("Predicted Result",predlr)
```

```python
print("Actual",y_test)
```

```python
print("Error")
print(mean_squared_error(y_test,predlr))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,predlr))
```
