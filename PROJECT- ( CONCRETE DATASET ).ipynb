```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=10)
```

```python
lm=LinearRegression()
```

```python
lm.score(x_train,y_train)
```

```python
pred=lm.predict(x_test)
```

```python
print("Predicted Result",pred)
```

```python
print("Error")
print(mean_squared_error(y_test,pred))
```
