# SPLITTING INTO X AND Y

```python
x=wine.data
y=wine.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.24,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# APPLYING MODELS 

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error,r2_score
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predlr=lr.predict(x_test)
```

```python
print(lr.score(x_train,y_train))
```

```python
print(r2_score(y_test,predlr))
```

```python
print(mean_squared_error(y_test,predlr))
```

# PIPELINE

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
pipeline=make_pipeline(RandomForestRegressor(n_estimators=200))
```

```python
pipeline.get_params()
```

```python
hyperparameters = { 'randomforestregressor__max_features' : ['auto','sqrt'],
                  'randomforestregressor__max_depth': [None,1,2,3]}
```

# GRIDSEARCHCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
gscv=GridSearchCV(pipeline,hyperparameters,cv=10)
gscv.fit(x_train,y_train)
predgscv=gscv.predict(x_test)
print(gscv.score(x_train,y_train))
print(r2_score(y_test,predgscv))
print(mean_squared_error(y_test,predgscv))
```

# SAVING THE MODEL

```python
from sklearn.externals import joblib
joblib.dump(gscv,"Wine.pkl")
```
